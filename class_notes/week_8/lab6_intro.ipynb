```python
# imports
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# user input
print("Your color maps options are: Greys, gist_earth, ocean or terrain")
print("")
c_map_input = input("Please select a color map: \n")
grid_input = input("\nDo you want grid lines? (yes or no): \n")
tick_label_input = input("\nDo you want tick labels? (yes or no): \n")
```

```python
# logic

# make changes to grid
if grid_input == 'yes':
    grid_option = True
elif grid_input == 'no':
    grid_option = False
else:
    grid_option = True

# select a color map
if c_map_input not in ['Greys','gist_earth','ocean','terrain']:
    cmap_choice = "terrain"
else:
    cmap_choice = c_map_input

# below, deal with the tick labels
```

```python
# code
image = plt.imread('stinkbug.png')
fig, ax = plt.subplots(figsize=(9,9))
ax.imshow(image, cmap=cmap_choice)

if tick_label_input == 'no':
    ax.set_xticklabels([])
    ax.set_yticklabels([])
else:
    pass
ax.grid(grid_option)
plt.show()
```
