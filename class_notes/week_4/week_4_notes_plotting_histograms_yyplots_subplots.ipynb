## Histograms

```python
# imports
import numpy as np
import matplotlib.pyplot as plt
# if using a Jupyter notebook:
%matplotlib inline

# create and manipulate data
mu = 75      # mean
sigma = 10   # standard deviation
x = np.random.normal(mu, sigma, size=200)

# create fig and ax objects
fig, ax = plt.subplots()

# plot data
ax.hist(x, 20)    # x is the array of numbers, 20 is the number of bins, # of bars on histogram

# customize plot
ax.set_title('Histogram with 20 bins')
ax.set_xlabel('bin range')
ax.set_ylabel('frequency')

# save and show the plot
plt.savefig('historgram.png', dpi=72)
plt.show()
```

## YY Plot

yy-plot is type of plot with two different y-axes, often each y-axis has a differnt scale, but the data on the yy-plot shares the same x-axis

```python
# imports
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# create and manipulate the data
x = np.arange(0,5,0.1)   # start 0, stop 5, count by 0.01
y1 = np.sin(x)
y2 = np.exp(x)

# create fig and ax objects
fig, ax1 = plt.subplots()

# plot the data on the left-hand y-axis and customize
ax1.plot(x,y1,'r')
ax1.set_xlabel('x from 0 to 5')
ax1.tick_params(axis='y', labelcolor='r')
ax1.set_ylabel('sin(x)')
ax1.set_title('yy-plot of sin(x) and exp(x)')

# plot the data on the right-hand y-axis and customize
ax2 = ax1.twinx()
ax2.plot(x,y2,'b')
ax2.tick_params(axis='y', labelcolor='b')
ax2.set_ylabel('exp(x)')

# save and show the plot
plt.savefig('yy-plot.png', dpi=300)
plt.show()
```

## In-Class Example Plot

```python
# imports
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# data
r = np.arange(0.01,1.01,0.01)
Ea = -1.436/r
Er = 7.32e-6/r**8
En = Ea + Er

# fig, ax objects
fig, ax = plt.subplots()

# plot data
ax.plot(r,Ea,'b')
ax.plot(r,Er,'r')
ax.plot(r,En,'y')

# customize plot
ax.set_xlabel('Interatomic Distance r (nm)')
ax.set_ylabel('Bonding Energy E (eV)')
ax.set_title('Bonding Energy vs. Interatomic Distance of an Na+ Cl- pair')
ax.set_xlim([0,1.00])
ax.set_ylim([-6,6])
ax.legend(['Ea','Er','En'])

# save and show the plot
plt.savefig('bond_energy_vs_distance.png', dpi=72)
plt.show()
```

## Subplots

Subplots are multiple plots in the same figure window. These subplots can be organized in rows and columns. Sometimes you need to be careful with the figure size when building subplots in order to get the aspect ratios you want in the subplots.

### Build 4 subplots with a linear, logx, logy, log-log

```python
# imports
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# data
x = np.arange(0,10,0.1)
y = np.exp(x)

# fig, multiple subplots
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(9,6))

# plot the data on each subplot and customize each subplot
# linear x, linear y
ax1.plot(x,y)
ax1.set_title('linear x and linear y')

# linear x, log y
ax2.semilogy(x,y)
ax2.set_title('linear x and log y')

# linear y, log x
ax3.semilogx(x,y)
ax3.set_title('log x and linear y')

# log x, log y
ax4.loglog(x,y)
ax4.set_title('log x and log y')

# save and show the figure with all subplots in it.
fig.tight_layout()
plt.savefig('4-subplots.png', dpi=72)
plt.show()
```
